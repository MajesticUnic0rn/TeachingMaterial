```python
# Assuming 'income' and 'marital_status' are columns in your DataFrame 'df'
# First, create a cross-tabulation of marital status by income level
income_distribution = pd.crosstab(input_df['w3_ppmarit'], input_df['w3_ppincimp'])

# Plotting
income_distribution.plot(kind='bar', stacked=False, figsize=(10, 6))
plt.title('Income Distribution by Marital Status')
plt.xlabel('Marital Status')
plt.xticks(rotation=45)
plt.ylabel('Frequency')
plt.legend(title='Income', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
