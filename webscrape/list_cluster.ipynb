```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Your data
x = np.array([[36, 7000], [45, 6000], [70, 7100], [20, 2000], [25, 2200], [15, 1800]])  # age and salary

# KMeans clustering
kmeans = KMeans(n_clusters=2)
kmeans.fit(x)

# Cluster centers
centroids = kmeans.cluster_centers_

# Cluster labels
labels = kmeans.labels_

# Plotting
plt.figure(figsize=(10, 6))
colors = ["blue", "green"]

# Plot each data point
for i in range(len(x)):
    plt.scatter(x[i][0], x[i][1], c=colors[labels[i]], s=50)

# Plot centroids
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=150, c='r')

plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('KMeans Clustering Visualization')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets as datasets
input_iris=datasets.load_iris()
```

```python
input_df = pd.DataFrame(data=input_iris.data,  
                  columns=input_iris.feature_names) 
# If you want to add the target (species) to the DataFrame
input_df['species'] = input_iris.target
# Optionally, map the numeric species to their original names
input_df['species'] = input_df['species'].map(dict(enumerate(input_iris.target_names)))
```

```python
input_df
```

```python
x=input_df.iloc[:,0:4]
```

```python
for i in range (1,6):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(x)
    input_df['cluster']=kmeans.predict(x)
    sns.pairplot(input_df, hue='cluster')
    plt.show()
```

```python
WCCS=[]
```

```python
for i in range (1,6):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(x)
    WCCS.append(kmeans.inertia_)
plt.plot(range(1,6),WCCS)
plt.xlabel('Number of clusters')
plt.ylabel('WCCS')
plt.show()
```
